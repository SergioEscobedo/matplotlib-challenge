```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "/Users/sergioescobedo/Desktop/matplotlib-challenge/Starter_Code 4/Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "/Users/sergioescobedo/Desktop/matplotlib-challenge/Starter_Code 4/Pymaceuticals/data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

merged = pd.merge(mouse_metadata, study_results, on='Mouse ID')

merged.head()
```

```python
grouped_regimen = merged.groupby('Drug Regimen')

statistics = grouped_regimen['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])

statistics.columns = ['Mean Tumor Volume', 'Median Tumor Volume', 'Tumor Volume Variance', 'Tumor Volume Std. Dev.', 'Tumor Volume Std. Err.']

statistics.reset_index(inplace=True)

(statistics)
```

```python
#Variables def
grouped_df = merged.groupby('Mouse ID')
max_timepoint = grouped_df['Timepoint'].max()
last_timepoint_df = pd.DataFrame({'Timepoint': max_timepoint}).reset_index()
final_volume_df = pd.merge(last_timepoint_df, merged, on=['Mouse ID', 'Timepoint'], how='left')
treatment_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# List
tumor_volume_data = []

# For cycle
for treatment in treatment_list:
   
    treatment_rows = final_volume_df[final_volume_df['Drug Regimen'] == treatment]
    
    tumor_volume_data.append(treatment_rows['Tumor Volume (mm3)'])

for i in range(len(treatment_list)):
    treatment = treatment_list[i]
    tumor_volumes = tumor_volume_data[i]
    
    quartiles = tumor_volumes.quantile([0.25, 0.5, 0.75])
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    iqr = upper_q - lower_q
    

    lower_bound = lower_q - 1.5 * iqr
    upper_bound = upper_q + 1.5 * iqr
    outliers = tumor_volumes.loc[(tumor_volumes < lower_bound) | (tumor_volumes > upper_bound)]
    
    print(f"Treatment: {treatment}")
    print(f"Quartiles: {quartiles}")
    print(f"IQR: {iqr}")
    print(f"Potential outliers: {outliers}\n")


```

```python
fig, ax = plt.subplots()

outlier_design = dict(marker='*', markerfacecolor='red', markersize=10)


ax.boxplot(tumor_volume_data, labels=treatment_list, flierprops=outlier_design)


ax.set_title('Final Tumor Volume Distribution by Treatment Group')
ax.set_xlabel('Treatment')
ax.set_ylabel('Tumor Volume (mm3)')

plt.show()
```
